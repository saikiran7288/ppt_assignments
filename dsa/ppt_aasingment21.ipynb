You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorder_traversal(node, values):
    if node is None:
        return
    inorder_traversal(node.left, values)
    values.append(node.value)
    inorder_traversal(node.right, values)

def convertToBST(root):
    if root is None:
        return None

    values = []
    inorder_traversal(root, values)
    values.sort()

    def inorder_modify(node):
        nonlocal values
        if node is None:
            return
        inorder_modify(node.left)
        node.value = values.pop(0)
        inorder_modify(node.right)

    inorder_modify(root)
    return root
```

```python
# Input binary tree
#         10
#        /  \
#       2    7
#      / \
#     8   4
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert binary tree to binary search tree
converted_root = convertToBST(root)

# Inorder traversal of converted binary search tree
def inorder_traversal(node):
    if node is None:
        return
    inorder_traversal(node.left)
    print(node.value, end=" ")
    inorder_traversal(node.right)

inorder_traversal(converted_root)  # Output: 2 4 7 8 10
```

```python
<aside>
ðŸ’¡ Question-2:

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Consider the following BST:

![1.png](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/f2455039-7e12-43fc-a7d3-b5be24931c1c/1.png)

**Input-1:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 6

node-2 = 14

**Output-1:**

The distance between the two keys = 4

**Input-2:**

n = 9

values = [8, 3, 1, 6, 4, 7, 10, 14,13]

node-1 = 3

node-2 = 4

**Output-2:**

The distance between the two keys = 2

</aside>

```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def findLowestCommonAncestor(root, node1, node2):
    if root is None:
        return None

    # If both nodes are smaller, LCA must be in the left subtree
    if root.value > node1 and root.value > node2:
        return findLowestCommonAncestor(root.left, node1, node2)

    # If both nodes are greater, LCA must be in the right subtree
    if root.value < node1 and root.value < node2:
        return findLowestCommonAncestor(root.right, node1, node2)

    # LCA found
    return root

def findDistanceFromNode(root, node, distance):
    if root is None:
        return -1

    # If node found, return the distance
    if root.value == node:
        return distance

    # Recursively search in left and right subtrees
    left_distance = findDistanceFromNode(root.left, node, distance + 1)
    if left_distance != -1:
        return left_distance

    right_distance = findDistanceFromNode(root.right, node, distance + 1)
    if right_distance != -1:
        return right_distance

    # Node not found in both subtrees
    return -1

def findDistanceBetweenNodes(root, node1, node2):
    # Find the lowest common ancestor (LCA) of the two nodes
    lca = findLowestCommonAncestor(root, node1, node2)

    # Calculate the distance from LCA to each node
    distance1 = findDistanceFromNode(lca, node1, 0)
    distance2 = findDistanceFromNode(lca, node2, 0)

    # Return the total distance
    return distance1 + distance2

# Create the BST
root = TreeNode(8)
root.left = TreeNode(3)
root.right = TreeNode(10)
root.left.left = TreeNode(1)
root.left.right = TreeNode(6)
root.left.right.left = TreeNode(4)
root.left.right.right = TreeNode(7)
root.right.right = TreeNode(14)
root.right.right.left = TreeNode(13)

# Example usage
print(findDistanceBetweenNodes(root, 6, 14))  # Output: 4
print(findDistanceBetweenNodes(root, 3, 4))   # Output: 2
```

```python
<aside>
ðŸ’¡ Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.prev = None
        self.next = None

def convertToDoublyLinkedList(root):
    if root is None:
        return None

    # Stack for iterative in-order traversal
    stack = []
    current = root
    prev = None
    head = None

    # Perform in-order traversal
    while current is not None or stack:
        while current is not None:
            stack.append(current)
            current = current.left

        current = stack.pop()

        # Update the pointers
        if prev is None:
            head = current
        else:
            prev.right = current
            current.prev = prev

        prev = current
        current = current.right

    return head

# Create the binary tree
root = Node(10)
root.left = Node(5)
root.right = Node(20)
root.right.left = Node(30)
root.right.right = Node(35)

# Convert the binary tree to a doubly linked list
head = convertToDoublyLinkedList(root)

# Print the doubly linked list
current = head
while current is not None:
    print(current.value, end=" ")
    current = current.right
```

```python
<aside>
ðŸ’¡ Question-4:

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

4 â†’ 5

5 â†’ 6

6 â†’ 7

7 â†’ -1

</aside>
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None

def connectNodesAtSameLevel(root):
    if root is None:
        return None

    # Perform level order traversal using a queue
    queue = [root]
    while queue:
        level_size = len(queue)

        # Connect nodes at the current level
        for i in range(level_size):
            node = queue.pop(0)

            # Set the next pointer
            if i < level_size - 1:
                node.next = queue[0]
            else:
                node.next = None

            # Add the child nodes to the queue
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return root

# Create the binary tree
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Connect nodes at the same level
connectNodesAtSameLevel(root)

# Print the connections
current = root
while current:
    node = current
    while node:
        if node.next:
            print(node.value, end=" â†’ ")
        else:
            print(node.value, end=" â†’ -1 ")
        node = node.next
    print()
    current = current.left
```
