Given a linked list ofÂ **N**Â nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**

```
Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.

```

**Example 2:**

```
Input:
N = 4
value[] = {1,8,3,4}
X = 0
Output:1
Explanation:TheÂ Linked list does not
contains any loop.
```

**Example 3:**
Input:
N = 4
value[] = {1,2,3,4}
X = 1
Output:1
Explanation:The link list looks like
1 -> 2 -> 3 -> 4
^              |
|______________|
A loop is present.
If you remove it successfully,
the answer will be 1.
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detect_and_remove_loop(head):
    # Check if the linked list has a loop
    slow = head
    fast = head

    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break

    # If there is no loop, return the original head
    if fast is None or fast.next is None:
        return head

    # Find the meeting point of the loop
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    # Unlink the last node from the loop
    fast.next = None

    return head

# Create the linked list: 1 -> 3 -> 4 -> 2 -> 3 (loop)
node1 = Node(1)
node2 = Node(3)
node3 = Node(4)
node4 = Node(2)
node5 = Node(3)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node3  # Create a loop

# Detect and remove the loop
new_head = detect_and_remove_loop(node1)

# Print the modified list
current = new_head
while current is not None:
    print(current.data, end=" ")
    current = current.next
print()
````

ðŸ’¡ **Question 2**

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**

```
Input:
LinkedList: 4->5->6
Output:457

```

**Example 2:**
Input:
LinkedList: 1->2->3
Output:124
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_linked_list(head):
    prev = None
    current = head

    while current is not None:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def add_one(head):
    head = reverse_linked_list(head)

    current = head
    carry = 1

    while current is not None:
        sum = current.data + carry

        if sum < 10:
            current.data = sum
            carry = 0
        else:
            current.data = sum % 10
            carry = 1

        prev = current
        current = current.next

    if carry == 1:
        new_node = Node(1)
        prev.next = new_node

    head = reverse_linked_list(head)

    return head

def print_linked_list(head):
    current = head

    while current is not None:
        print(current.data, end="->")
        current = current.next

    print("None")

# Create the linked list
node1 = Node(4)
node2 = Node(5)
node3 = Node(6)

node1.next = node2
node2.next = node3

print("Original Linked List:")
print_linked_list(node1)

# Add 1 to the linked list
new_head = add_one(node1)

print("Modified Linked List:")
print_linked_list(new_head)
````

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) aÂ **next**Â pointer to the next node,(ii) aÂ **bottom**Â pointerÂ to a linked list where this node is head.Each of theÂ sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order.Â **Note:**Â The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

```
Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output:Â 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```

**Example 2:**
Input:
5 -> 10 -> 19 -> 28
|          |
7          22
|          |
8          50
|
30
Output: 5->7->8->10->19->22->28->30->50
Explanation:
The resultant linked lists has every
node in a single level.

(Note:| represents the bottom pointer.)
</aside>

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def merge_sorted_lists(a, b):
    # Base cases
    if a is None:
        return b
    if b is None:
        return a

    # Compare the nodes and merge them recursively
    if a.data <= b.data:
        result = a
        result.bottom = merge_sorted_lists(a.bottom, b)
    else:
        result = b
        result.bottom = merge_sorted_lists(a, b.bottom)

    return result

def flatten_linked_list(head):
    # Base case: an empty list or a single node
    if head is None or head.next is None:
        return head

    # Recursively flatten the rest of the list
    head.next = flatten_linked_list(head.next)

    # Merge the current list with the flattened list
    head = merge_sorted_lists(head, head.next)

    return head

def print_list(head):
    current = head
    while current is not None:
        print(current.data, end="->")
        current = current.bottom
    print("None")

# Create the linked list
node1 = Node(5)
node2 = Node(10)
node3 = Node(19)
node4 = Node(28)
node5 = Node(7)
node6 = Node(20)
node7 = Node(22)
node8 = Node(35)
node9 = Node(8)
node10 = Node(50)
node11 = Node(30)
node12 = Node(45)

# Set the next pointers
node1.next = node2
node2.next = node3
node3.next = node4
node5.next = node6
node6.next = node7
node9.next = node10
node11.next = node12

# Set the bottom pointers
node1.bottom = node5
node2.bottom = node9
node3.bottom = node11
node6.bottom = node10
node7.bottom = node12

# Flatten the linked list
flattened_head = flatten_linked_list(node1)

# Print the flattened list
print_list(flattened_head)
````

Construct a copy of the given list. The copy should consist of exactlyÂ **N**Â new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodesÂ **X**Â andÂ **Y**Â in the original list, whereÂ **X.arb**Â **-->**Â **Y**, then for the corresponding two nodesÂ **x**Â andÂ **y**Â in the copied list,Â **x.arb --> y.**

Return the head of the copied linked list.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def print_linked_list(head):
    current = head

    while current is not None:
        random_data = None if current.random is None else current.random.data
        print("Data:", current.data, "Random Data:", random_data)
        current = current.next

def clone_linked_list(head):
    if head is None:
        return None

    # Step 1: Create a copy of each node and insert it between the original node and the next node
    current = head

    while current is not None:
        new_node = Node(current.data)
        new_node.next = current.next
        current.next = new_node
        current = new_node.next

    # Step 2: Set the random pointer of the copied node to the corresponding random node in the original list
    current = head

    while current is not None:
        if current.random is not None:
            current.next.random = current.random.next
        current = current.next.next

    # Step 3: Separate the original and copied nodes into two separate linked lists
    original_head = head
    copied_head = head.next
    copied_current = copied_head

    while original_head is not None:
        original_head.next = original_head.next.next
        copied_current.next = copied_current.next.next
        original_head = original_head.next
        copied_current = copied_current.next

    # Step 4: Restore the original linked list by fixing the next pointers of the original nodes and the copied nodes
    return copied_head

# Create the original linked list
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

node1.next = node2
node2.next = node3
node3.next = node4

node1.random = node3
node2.random = node1
node3.random = node4
node4.random = node2

# Clone the linked list
cloned_head = clone_linked_list(node1)

# Print the cloned linked list
print("Cloned Linked List:")
print_linked_list(cloned_head)
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def print_linked_list(head):
    current = head

    while current is not None:
        print(current.val, end=" ")
        current = current.next

    print()

def odd_even_linked_list(head):
    if head is None or head.next is None:
        return head

    oddHead = head
    evenHead = head.next
    odd = oddHead
    even = evenHead

    while even is not None and even.next is not None:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = evenHead

    return oddHead

# Create the original linked list
head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Group the nodes by odd and even indices
reordered_head = odd_even_linked_list(head)

# Print the reordered linked list
print("Reordered Linked List:")
print_linked_list(reordered_head)
````

````python
<aside>
ðŸ’¡ **Question 6**

Given a singly linked list of sizeÂ **N**. The task is toÂ **left-shift**Â the linked list byÂ **k**Â nodes, whereÂ **k**Â is a given positive integer smaller than or equal to length of the linked list.

**Example 1:**

```
Input:
N = 5
value[] = {2, 4, 7, 8, 9}
k = 3
Output:8 9 2 4 7
Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2
Rotate 2:Â 7Â -> 8Â -> 9Â -> 2Â -> 4
Rotate 3:Â 8Â -> 9Â -> 2Â -> 4Â -> 7

```

**Example 2:**
Input:
N = 8
value[] = {1, 2, 3, 4, 5, 6, 7, 8}
k = 4
Output:5 6 7 8 1 2 3 4
</aside>
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def print_linked_list(head):
    current = head

    while current is not None:
        print(current.val, end=" ")
        current = current.next

    print()

def left_shift_linked_list(head, k):
    if head is None or k == 0:
        return head

    # Find the (k+1)th node from the beginning of the list
    current = head
    count = 1

    while count < k and current is not None:
        current = current.next
        count += 1

    if current is None:
        return head

    # Set the new head of the shifted list
    new_head = current.next

    # Connect the end of the list with the original head
    current.next = None

    # Traverse to the end of the list
    current = new_head

    while current.next is not None:
        current = current.next

    # Connect the end of the list with the original head
    current.next = head

    return new_head

# Create the original linked list
head = ListNode(2)
node2 = ListNode(4)
node3 = ListNode(7)
node4 = ListNode(8)
node5 = ListNode(9)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Left-shift the linked list by k=3 nodes
shifted_head = left_shift_linked_list(head, 3)

# Print the shifted linked list
print("Shifted Linked List:")
print_linked_list(shifted_head)
````

````python
<aside>
ðŸ’¡ **Question 7**

You are given theÂ `head`Â of a linked list withÂ `n`Â nodes.

For each node in the list, find the value of theÂ **next greater node**. That is, for each node, find the value of the first node that is next to it and has aÂ **strictly larger**Â value than it.

Return an integer arrayÂ `answer`Â whereÂ `answer[i]`Â is the value of the next greater node of theÂ `ith`Â node (**1-indexed**). If theÂ `ith`Â node does not have a next greater node, setÂ `answer[i] = 0`.

**Example 1:**

!https://assets.leetcode.com/uploads/2021/08/05/linkedlistnext1.jpg

```
Input: head = [2,1,5]
Output: [5,5,0]

```

**Example 2:**

Input: head = [2,7,4,3,5]
Output: [7,0,5,5,0]
</aside>
````

````python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def next_greater_nodes(head):
    # Step 1: Store the values of each node in a list
    values = []
    current = head

    while current is not None:
        values.append(current.val)
        current = current.next

    # Step 2: Initialize stack and result list
    stack = []
    result = [0] * len(values)

    # Step 3: Iterate over the values in reverse order
    for i in range(len(values)-1, -1, -1):
        # Step 4: Pop elements from stack if necessary
        while stack and stack[-1] <= values[i]:
            stack.pop()

        # Step 5: Set result[i] as 0 if stack is empty
        if not stack:
            result[i] = 0
        # Step 6: Set result[i] as top element of stack
        else:
            result[i] = stack[-1]

        # Step 7: Push current value onto stack
        stack.append(values[i])

    return result

# Create the linked list
head = ListNode(2)
node2 = ListNode(7)
node3 = ListNode(4)
node4 = ListNode(3)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Find the next greater nodes
result = next_greater_nodes(head)

# Print the result
print(result)
````
